```python
import numpy as np
import math
import scipy as sp
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
def load_data(dir):
  data = np.zeros((5,19,8))
  data[0,:,:] = np.loadtxt('/content/gdrive/MyDrive/' + dir + '/Data_0.csv',delimiter=',', skiprows=1)
  data[1,:,:] = np.loadtxt('/content/gdrive/MyDrive/' + dir + '/Data_15.csv',delimiter=',', skiprows=1)
  data[2,:,:] = np.loadtxt('/content/gdrive/MyDrive/' + dir + '/Data_-15.csv',delimiter=',', skiprows=1)
  data[3,:,:] = np.loadtxt('/content/gdrive/MyDrive/' + dir + '/Data_30.csv',delimiter=',', skiprows=1)
  data[4,:,:] = np.loadtxt('/content/gdrive/MyDrive/' + dir + '/Data_-30.csv',delimiter=',', skiprows=1)
  return data
```

```python
data1 = load_data("Yifan_7_23_data_#1")
data2 = load_data("Yifan_7_23_data_#2")
data3 = load_data("Yifan_7_23_data_#3")
```

```python
def plotter(ecc):
  if ecc == 0:
    ind = 0
  elif ecc == 15:
    ind = 1
  elif ecc == -15:
    ind = 2
  elif ecc == 30:
    ind = 3
  else:
    ind = 4
  x = data1[ind,:,0]
  y1 = -data1[ind,:,5]
  y2 = -data2[ind,:,5]
  y3 = -data3[ind,:,5]
  z1 = -data1[ind,:,1]
  z2 = -data2[ind,:,1]
  z3 = -data3[ind,:,1]
  colors1 = ['red' if i == 0 else 'green' if i == -8 else 'blue' if i == 8 else 'yellow' if i < -17 else 'purple' for i in z1]
  colors2 = ['red' if i == 0 else 'green' if i == -8 else 'blue' if i == 8 else 'yellow' if i < -17 else 'purple' for i in z2]
  colors3 = ['red' if i == 0 else 'green' if i == -8 else 'blue' if i == 8 else 'yellow' if i < -17 else 'purple' for i in z3]
  plt.scatter(x, y1, c=colors1)
  plt.scatter(x, y2, c=colors2)
  plt.scatter(x, y3, c=colors3)
  # plt.plot(x, z, label='Target position')
  plt.title('Eccentricity is ' + str(ecc) + ' degrees')
  plt.xlim((0,60))
  plt.xlabel('Time (s)')
  plt.ylabel('Angular disparity (degrees)')
  red_dot_legend = plt.Line2D([0], [0], marker='o', color='red', linestyle='None', markersize=5, label='Target position 0')
  green_dot_legend = plt.Line2D([0], [0], marker='o', color='green', linestyle='None', markersize=5, label='Target position -15')
  blue_dot_legend = plt.Line2D([0], [0], marker='o', color='blue', linestyle='None', markersize=5, label='Target position 15')
  yellow_dot_legend = plt.Line2D([0], [0], marker='o', color='yellow', linestyle='None', markersize=5, label='Target position -30')
  purple_dot_legend = plt.Line2D([0], [0], marker='o', color='purple', linestyle='None', markersize=5, label='Target position 30')
  plt.legend(handles=[red_dot_legend, green_dot_legend, blue_dot_legend, yellow_dot_legend, purple_dot_legend])
  plt.show()
```

```python
plotter(0)
```

```python
plotter(15)
```

```python
plotter(-15)
```
